## LDA
    (Linear Discriminant Analysis)是一种经典的线性判别方法，又称Fisher判别分析。该方法思想比较简单:给定训练集样例，设法将样例投影到一维的直线上，使得同类样例的投影点尽可能接近和密集（即希望类内离散度越小越好），异类投影点尽可能远离（即希望两类的均值点之差越小越好）
https://blog.csdn.net/winycg/article/details/83043663

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def lda3d():
    mat1 = np.loadtxt('boy.txt', dtype=np.float32)
    b = mat1
    mat0 = np.loadtxt('girl.txt', dtype=np.float32)
    g = mat0

    # train
    # 均值u,各类离散度矩阵
    u1, sgm1 = getUSgm(b)
    u0, sgm0 = getUSgm(g)
    # 总的类内离散度矩阵
    sw = sgm1 + sgm0
    # 类间离散度矩阵
    cha = u1 - u0
    sb = cha.T * cha
    # 计算权向量(乘完是一个列向量)
    w = sw.I * cha.T
    # 计算偏置
    u11 = w.T * u1.T
    u00 = w.T * u0.T
    w0 = (-1.0 / 2) * (float(u11) + float(u00))

    errTest(w0,w)

    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(b[:, 0], b[:, 1], b[:, 2], c='b',label='male')
    ax.scatter(g[:, 0], g[:, 1], g[:, 2], c='g', label='female')

    # 绘制决策面
    X = np.arange(140, 210, 10)  # 采样
    Y = np.arange(30, 100, 10)
    X, Y = np.meshgrid(X, Y)  # 笛卡尔积
    w1 = float(w[0, 0])
    w2 = float(w[1, 0])
    w3 = float(w[2, 0])
    Z = -(w0 + w1 * X + w2 * Y) / w3
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow',alpha = 0.8)
    ax.set_xlabel('Height', color='r')
    ax.set_ylabel('Weight', color='g')
    ax.set_zlabel('Shoe Size', color='b')
    plt.show()
    getROC(w0,w)
```

```python
#测试,传入的w是列向量!
def errTest(w0,w):
    mat1 = np.loadtxt('boynew.txt', dtype=np.float32)
    b = mat1

    mat0 = np.loadtxt('girlnew.txt', dtype=np.float32)
    g = mat0

    errCount = 0
    for i in b:
        if i*w+w0<0:
            errCount+=1
    for i in g:
        if i*w+w0>0:
            errCount+=1
    print("After test for ",len(b)+len(g)," times, and the error rate is ",errCount/(len(b)+len(g)))
```

```python
def getROC(w0,w):
    X=[]
    Y=[]
    # 分类面采样阈值从w0-0.1到w0+0.1,步长0.001
    # 但range只能用int类型,所以乘以1000倍,传入时再除以1000
    for b in range(int(1000 * w0 - 100), int(1000 * w0 + 100), 1):
        # 获取假阳率和真阳率
        FPR, RECALL = getFprRecall(float(b) / 1000, w)
        # 分别加入到坐标采样列表中
        X.append(FPR)
        Y.append(RECALL)
    # 绘制ROC曲线
    plt.plot(X, Y, "g-", linewidth=2)
    # 横纵坐标名称,标题名称
    plt.xlabel('FPR')
    plt.ylabel('RECALL')
    plt.show()
```

```python
#获取假阳性率FPR和召回率RECALL
def getFprRecall(w0,w):
    mat1 = np.loadtxt('boynew.txt', dtype=np.float32)
    b = mat1

    mat0 = np.loadtxt('girlnew.txt', dtype=np.float32)
    g = mat0
    
    errPstv = 0.0  # 记录假阳性样本数
    relPstv = 0.0  # 记录真阳性样本数
    for i in b:
        if i*w+w0>0:
            relPstv+=1
    for i in g:
        if i*w+w0>0:
            errPstv+=1
    return errPstv / len(g), relPstv / len(b)
```

```python
def getUSgm(xMat):# 均值
    # 行数1,列数xMat的列数
    u = np.zeros((1, xMat.shape[1]))
    u = np.mat(u)
    for vec in xMat:
        u += vec
    u /= xMat.shape[0]  # 除以行数
    sgm = np.zeros((xMat.shape[1], xMat.shape[1]))
    sgm = np.mat(sgm)
    for vec in xMat:
        cha = vec - u
        sgm += cha.T * cha
    return u, sgm
```

```python
def fisherBayles(i=0,j=1):
    def gauss2(x, miu, sgm):
        r = np.mat([x[0] - miu[0], x[1] - miu[1]])
        multi = r * sgm.I * r.T
        multi = float(multi)  # 1乘1矩阵取内容 I求逆,.T转置
        k = np.exp(-multi / 2)
        k /= 2 * np.math.pi * (abs(np.linalg.det(sgm)) ** (1 / 2))  # linalg.det求行列式的值
        return k

    def getData():
        fr1 = open(r'boynew.txt')
        fr0 = open(r'girlnew.txt')
        arrayOLines1 = fr1.readlines()  # 读取文件
        arrayOLines0 = fr0.readlines()
        dataMat1 = []
        dataMat0 = []
        for line in arrayOLines1:
            line = line.strip()  # strip()去掉首尾空格
            listFromLine = line.split()
            for i in range(len(listFromLine)):
                listFromLine[i] = float(listFromLine[i])
            dataMat1.append(listFromLine)
        for line in arrayOLines0:
            line = line.strip()  # strip()去掉首尾空格
            listFromLine = line.split()
            for i in range(len(listFromLine)):
                listFromLine[i] = float(listFromLine[i])
            dataMat0.append(listFromLine)
        return dataMat1, dataMat0

    def TraBys(i, j):
        dataMat1, dataMat0 = getData()
        dataMat1 = np.mat(dataMat1)
        dataMat0 = np.mat(dataMat0)

        miu1, miu0, sgm1, sgm0 = getMiuSgm(i, j, dataMat1, dataMat0)
        return miu1, miu0, sgm1, sgm0

    def getMiuSgm(i, j, dataMat1, dataMat0):
        Imiu1 = 0.0
        for r in range(len(dataMat1)):
            Imiu1 += dataMat1[r, i]
        Imiu1 = Imiu1 / len(dataMat1)
        Imiu0 = 0.0
        for r in range(len(dataMat0)):
            Imiu0 += dataMat0[r, i]
        Imiu0 = Imiu0 / len(dataMat0)
        Jmiu1 = 0.0
        for r in range(len(dataMat1)):
            Jmiu1 += dataMat1[r, j]
        Jmiu1 = Jmiu1 / len(dataMat1)
        Jmiu0 = 0.0
        for r in range(len(dataMat0)):
            Jmiu0 += dataMat0[r, j]
        Jmiu0 = Jmiu0 / len(dataMat0)
        miu1 = [Imiu1, Jmiu1]
        miu0 = [Imiu0, Jmiu0]

        sgm1 = np.mat(np.zeros((2, 2)))
        for k in range(len(dataMat1)):
            x = np.mat([dataMat1[k, i] - miu1[0], dataMat1[k, j] - miu1[1]])
            sgm1 += x.T * x
        sgm1 /= len(dataMat1)

        sgm0 = np.mat(np.zeros((2, 2)))
        for k in range(len(dataMat0)):
            x = np.mat([dataMat0[k, i] - miu0[0], dataMat0[k, j] - miu0[1]])
            sgm0 += x.T * x
        sgm0 /= len(dataMat0)
        return miu1, miu0, sgm1, sgm0
    # 返回两均值,两方差

    def f(sample_height, sample_weight, miu1, miu0, sgm1, sgm0):
        mytemp1 = np.zeros(shape=(100, 100))
        dettemp = (np.linalg.det(sgm0)) / (np.linalg.det(sgm1))
        sample_vector = np.matrix([[sample_height[0, 0]], [sample_weight[0, 0]]]).T
        # print( (sample_vector-miu0)* (np.linalg.inv(sgm0)))
        for i in range(100):
            for j in range(100):
                sample_vector = np.matrix([[sample_height[i, j]], [sample_weight[i, j]]])
                sample_vector_T = np.transpose(sample_vector)
                # 定义决策函数
                mytemp1[i, j] = 0.5 * (sample_vector_T - miu0) * (np.linalg.inv(sgm0)) * \
                                np.transpose(sample_vector_T - miu0) - 0.5 * (sample_vector_T - miu1) * \
                                (np.linalg.inv(sgm1)) * np.transpose(sample_vector_T - miu1) + 0.5 * np.math.log(dettemp)
        return mytemp1

    def Go(i, j):  # 演示用的函数,传入两个特征编号i和j
        miu1, miu0, sgm1, sgm0 = TraBys(i, j)
        sample_height = np.linspace(150, 180, 100)
        sample_weight = np.linspace(40, 80, 100)
        Sample_height, Sample_weight = np.meshgrid(sample_height, sample_weight)
        plt.contourf(Sample_height, Sample_weight, f(Sample_height, Sample_weight, miu1, miu0, sgm1, sgm0), 0, alpha=0)
        # 绘制等高线,圈内为女生，圈外为男生
        C = plt.contour(Sample_height, Sample_weight, f(Sample_height, Sample_weight, miu1, miu0, sgm1, sgm0), 0,
                        colors='black', linewidths=1.2)
        # # 显示各等高线的数据标签
        plt.clabel(C, inline=True, fontsize=10)
        plt.xlabel('height/cm')
        plt.ylabel('weight/kg')

    mat1 = np.loadtxt('boy.txt', dtype=np.float32)
    b = np.append(mat1.transpose()[i], mat1.transpose()[j], axis=0)
    b = np.reshape(b, (2, int(len(b) / 2)))
    b = b.T

    mat0 = np.loadtxt('girl.txt', dtype=np.float32)
    g = np.append(mat0.transpose()[i], mat0.transpose()[j], axis=0)
    g = np.reshape(g, (2, int(len(g) / 2)))
    g = g.T

    plt.scatter(b[:, 0], b[:, 1], c='#99CC99', label='male')
    plt.scatter(g[:, 0], g[:, 1], c='#FFCC00', label='female')
    # 均值u,各类离散度矩阵
    u1, sgm1 = getUSgm(b)
    u0, sgm0 = getUSgm(g)
    # 总的类内离散度矩阵
    sw = sgm1 + sgm0
    # 类间离散度矩阵
    cha = u1 - u0
    sb = cha.T * cha

    w = sw.I * cha.T  # 权向量(乘完是一个列向量)
    # print("权向量：", end=' ')
    # print(w.T)
    u11 = w.T * u1.T
    u00 = w.T * u0.T
    w0 = (-1.0 / 2) * (float(u11) + float(u00))  # 偏置
    # print("偏置：", end=' ')
    # print(w0)

    # 绘制曲线用的横坐标(i特征)
    if i == 0:
        lft = 140
        rgt = 200
        xlab = 'height'
    elif i == 1:
        lft = 35
        rgt = 95
        xlab = 'weight'
    elif i == 2:
        lft = 30
        rgt = 47
        xlab = 'size'
    x1 = np.linspace(lft, rgt, 20)
    plt.xlim(lft, rgt)
    plt.xlabel(xlab)
    if j == 0:
        lft = 140
        rgt = 200
        ylab = 'height'
    elif j == 1:
        lft = 35
        rgt = 95
        ylab = 'weight'
    elif j == 2:
        lft = 30
        rgt = 47
        ylab = 'size'

    y1 = np.linspace(20, 80, 20)
    plt.ylabel(ylab)
    plt.ylim(lft, rgt)
    wi = float(w[0, 0])
    wj = float(w[1, 0])
    y1 = (wj / wi) * x1
    # print(y1)
    plt.plot(x1, -(wi * x1 + w0) / wj, c=u'r')
    plt.plot(x1, y1, c='b')
    Go(0,1)
    plt.show()
```

```python
#留一法获取交叉验证错误率
def getCrossValidationOnly1():
    def getData(pathstr1, pathstr0):
        fr1 = open(pathstr1)
        fr0 = open(pathstr0)
        arrayOLines1 = fr1.readlines()  # 读取文件
        arrayOLines0 = fr0.readlines()
        dataMat1 = []
        dataMat0 = []
        for line in arrayOLines1:
            line = line.strip()  # strip()去掉首尾空格
            listFromLine = line.split()  # 按空白字符分割成列表
            for i in range(len(listFromLine)):
                listFromLine[i] = float(listFromLine[i])
            dataMat1.append(listFromLine)
        for line in arrayOLines0:
            line = line.strip()  # strip()去掉首尾空格
            listFromLine = line.split()
            for i in range(len(listFromLine)):
                listFromLine[i] = float(listFromLine[i])
            dataMat0.append(listFromLine)
        return dataMat1, dataMat0

    def Train(dataMat1, dataMat0):
        xMat1 = np.mat(dataMat1)
        xMat0 = np.mat(dataMat0)
        u1, sgm1 = getUSgm(xMat1)
        u0, sgm0 = getUSgm(xMat0)
        # 总的类内离散度矩阵
        sw = sgm1 + sgm0
        # 类间离散度矩阵
        cha = u1 - u0
        sb = cha.T * cha
        # 计算权向量(乘完是一个列向量)
        w = sw.I * cha.T
        # 计算偏置
        u11 = w.T * u1.T
        u00 = w.T * u0.T
        w0 = (-1.0 / 2) * (float(u11) + float(u00))
        return w0, w

    print("留一法交叉验证")
    errCount = 0.0
    # 遍历作留一法交叉验证
    dataMat1, dataMat0 = getData(r'.\boy.txt', r'.\girl.txt')
    for i in range(len(dataMat1)):
        subMat = dataMat1[0:i]
        subMat_End = dataMat1[i + 1:]
        subMat.extend(subMat_End)
        # 第i个 没了
        w0, w = Train(subMat, dataMat0)  # 训练
        xVec = np.mat(dataMat1[i])
        if xVec * w + w0 < 0:  # 对男生而言,<0是错误分类
            errCount += 1
    for i in range(len(dataMat0)):
        subMat = dataMat0[0:i]
        subMat_End = dataMat0[i + 1:]
        subMat.extend(subMat_End)
        # 第i个 没了
        w0, w = Train(dataMat1, subMat)  # 训练
        xVec = np.mat(dataMat0[i])
        if xVec * w + w0 > 0:  # 女 错误
            errCount += 1
    nth = len(dataMat1) + len(dataMat0)
    print(nth,"次留一法测试，错误率",errCount/nth)
```
