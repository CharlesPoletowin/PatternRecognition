```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def getData(filename="2.txt"):
    f = open(filename)
    dataMat=np.loadtxt(filename)

    return np.mat(dataMat)
```

```python
def MyPCA(dataMat, d):
    avgVec = np.mean(dataMat, axis=0)

    rmMat = dataMat - avgVec

    sigmaMat = np.cov(rmMat, rowvar=0)

    egnVal, egnVec = np.linalg.eig(np.mat(sigmaMat))

    egnValIdx = np.argsort(egnVal)

    egnValIdx = egnValIdx[:-d - 1:-1]

    newEngVec = egnVec[:, egnValIdx]

    lowDataMat = rmMat * newEngVec

    return lowDataMat, egnVal


def Show(myMat, k):
    plt.scatter(myMat[:, 0].flatten().A[0][0:k], \
                myMat[:, 1].flatten().A[0][0:k], \
                marker='.', s=50, c='blue', label='male')

    plt.scatter(myMat[:, 0].flatten().A[0][k:], \
                myMat[:, 1].flatten().A[0][k:], \
                marker='.', s=50, c='red', label='female')


def Mapping(egnVal):
    egnValIdx = np.argsort(egnVal)

    mysum = sum(egnVal)

    ax = plt.gca()
    ax.set_xticks(range(1, len(egnVal) + 1))

    X = range(1, len(egnVal) + 1)
    Y = (egnVal[egnValIdx] / mysum)[::-1]
    plt.plot(X, Y, "b", linewidth=1, marker='o')

    for a, b in zip(X, Y):
        plt.text(a, b + 0.02, '%f' % b, ha='center', \
                 va='bottom', fontsize=11, color='#BC13FE')

    plt.bar(X, Y, width=0.5, facecolor="#33CC99", edgecolor="white", yerr=0)
    plt.ylim(0, 1)
    ax.set_xlabel('Principal components')
    ax.set_ylabel('The percentage of its eigenvalues')
    plt.title('Eigenvalue Atlas of principal component analysis')
    plt.show()


def GoPCA():
    # 以k为男女分界,训练集测试集分界,一共4*k个样本
    # 男(训练),女(训练),男(测试),女(测试)各k个
    k = 50

    dataMat = getData()
    # 获取降维后的数据集和特征值数组
    allLowDataMat, engVal = MyPCA(dataMat, 2)
    lowDataMat = allLowDataMat[0:2 * k] #挑出前2k个当训练集
    oldDataMat = dataMat[0:2 * k]    #原始的训练集们
    newTstDataMat = allLowDataMat[2 * k:]#剩下2k个当测试集
    tstDataMat = dataMat[2 * k:]

    print("正在绘制PCA的本征值图谱..")
    # 利用特征值数组绘制主成分分析的本征值图谱
    Mapping(engVal)

    print("正在绘制降维后的图像..")
    # 绘制降维后的图像,传入男女分界点阈值
    # 样本数据'男|女|男|女'隔开了,不妨训练集和测试集分别绘制
    Show(lowDataMat, k)
    Show(newTstDataMat, k)
    plt.xlabel('New features of No.1')
    plt.ylabel('New features of No.2')
    plt.title('The new feature space')
    plt.show()
```

```python
GoPCA()
```

```python
def getUSgm(xMat):
    # 行数1,列数xMat的列数
    u = np.zeros((1, xMat.shape[1]))
    u = np.mat(u)
    for vec in xMat:
        u += vec
    u /= xMat.shape[0]
    sgm = np.zeros((xMat.shape[1], xMat.shape[1]))
    sgm = np.mat(sgm)
    for vec in xMat:
        cha = vec - u
        sgm += cha.T * cha
    return u, sgm


def Train(xMat1, xMat0):
    '''
    # 训练(共用)
    # 传入男女的mat对象,获取偏置和权向量
    '''
    # 均值u,各类离散度矩阵
    u1, sgm1 = getUSgm(xMat1)
    u0, sgm0 = getUSgm(xMat0)
    # 总的类内离散度矩阵
    sw = sgm1 + sgm0
    # 类间离散度矩阵
    cha = u1 - u0
    sb = cha.T * cha
    # 计算权向量(乘完是一个列向量)
    w = sw.I * cha.T
    # 计算偏置
    u11 = w.T * u1.T
    u00 = w.T * u0.T
    w0 = (-1.0 / 2) * (float(u11) + float(u00))
    return w0, w


def myFig(dataMat1, dataMat0, w0, w):
    '''
    绘制三维情况下的分类面和散点图
    '''
    x1 = [dataMat1[i, 0] for i in range(len(dataMat1))]
    y1 = [dataMat1[i, 1] for i in range(len(dataMat1))]
    z1 = [dataMat1[i, 2] for i in range(len(dataMat1))]
    x0 = [dataMat0[i, 0] for i in range(len(dataMat0))]
    y0 = [dataMat0[i, 1] for i in range(len(dataMat0))]
    z0 = [dataMat0[i, 2] for i in range(len(dataMat0))]
    # 绘制男女离散的点
    fig = plt.figure()
    ax = fig.add_subplot(121, projection='3d')

    ax.scatter(x1, y1, z1, c=u'b', label='male')
    ax.scatter(x0, y0, z0, c=u'r', label='female')
    # 绘制决策面
    X = np.arange(140, 210, 10)
    Y = np.arange(30, 100, 10)
    # 笛卡尔积
    X, Y = np.meshgrid(X, Y)
    w1 = float(w[0, 0])
    w2 = float(w[1, 0])
    w3 = float(w[2, 0])
    Z = -(w0 + w1 * X + w2 * Y) / w3
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
    # 给三个坐标轴注明
    ax.set_xlabel('Height', color='r')
    ax.set_ylabel('Weight', color='g')
    ax.set_zlabel('Shoe Size', color='b')


def oldShowfisher(oldMat,k):
    '''
    展示降维前的数据集
    '''
    xMat1,xMat0=oldMat[0:k],oldMat[k:]
    w0,w=Train(xMat1,xMat0)
    myFig(xMat1,xMat0,w0,w)
    plt.title('3D Fisher')
    return w0,w


def newShowfisher(newMat, k):
    '''
    #展示降维后的数据集
    '''
    xMat1, xMat0 = newMat[0:k], newMat[k:]
    # 男生
    plt.scatter(xMat1[:, 0].flatten().A[0], \
                xMat1[:, 1].flatten().A[0], \
                marker='.', s=50, c='blue', label='male')
    # 女生
    plt.scatter(xMat0[:, 0].flatten().A[0], \
                xMat0[:, 1].flatten().A[0], \
                marker='.', s=50, c='green', label='female')
    ax = plt.gca()
    ax.set_xlabel('New features of No.0')
    ax.set_ylabel('New features of No.1')
    plt.title('PCA-2D Fisher')
    w0, w = Train(xMat1, xMat0)

    # 绘制曲线用的横坐标(i特征)
    lft = -15
    rgt = 15
    x1 = np.linspace(lft, rgt, 10)
    # 绘制得到的直线
    wi = float(w[0, 0])
    wj = float(w[1, 0])
    plt.plot(x1, -(wi * x1 + w0) / wj, c=u'k')
    return w0, w


def getROC(w0, w, dataMat, k, colour):
    '''
    绘制ROC曲线,使用时传入不同的颜色
    '''
    X = []
    Y = []
    # 分类面采样阈值从w0-0.1到w0+0.1,步长0.001
    # 但range只能用int类型,所以乘以1000倍,传入时再除以1000
    for b in range(int(1000 * w0 - 100), int(1000 * w0 + 100), 1):
        # 获取假阳率和真阳率
        FPR, RECALL = getFprRecall(float(b) / 1000, w, dataMat, k)
        # 分别加入到坐标采样列表中
        X.append(FPR)
        Y.append(RECALL)
    # 绘制ROC曲线
    plt.plot(X, Y, color=colour, linewidth=2)

    plt.xlabel('FPR')
    plt.ylabel('RECALL')


def getFprRecall(w0, w, dataMat, k):
    '''
    #获取假阳性率FPR和召回率RECALL
    '''
    xMat1, xMat0 = dataMat[0:k], dataMat[k:]
    errPstv = 0.0
    relPstv = 0.0
    for line in xMat1:
        if line * w + w0 > 0:
            relPstv += 1
    for line in xMat0:
        if line * w + w0 > 0:
            errPstv += 1
    return errPstv / len(xMat0), relPstv / len(xMat1)


def GoFisher(oldMat, newMat, k):
    print("正在绘制Fisher分类面比较图..")

    old_w0, old_w = oldShowfisher(oldMat, k)
    plt.subplot(122)
    new_w0, new_w = newShowfisher(newMat, k)
    plt.show()

    print("正在绘制Fisher方法的ROC曲线..")

    getROC(old_w0, old_w, oldMat, k, '#CC6699')
    getROC(new_w0, new_w, newMat, k, '#9999FF')
    plt.legend(['3D Fisher', 'PCA-2D Fisher'])
    plt.xlabel('FPR')
    plt.ylabel('RECALL')
    plt.title('Fisher\'s ROC curve')
    plt.grid(True)
    plt.show()


def Gofisher2():
    k = 50
    dataMat = getData()
    # 获取降维后的数据集和特征值数组
    allLowDataMat, engVal = MyPCA(dataMat, 2)
    lowDataMat = allLowDataMat[0:2 * k]  # 挑出前2k个当训练集
    oldDataMat = dataMat[0:2 * k]  # 原始的训练集们
    GoFisher(oldDataMat,lowDataMat,k)
```

```python
Gofisher2()
```

```python
def gauss2(x, miu, sgm):
    r = np.mat([x[0]-miu[0], x[1]-miu[1]])
    multi = r * sgm.I * r.T
    multi = float(multi)  # 1乘1矩阵取内容 I求逆,.T转置
    k = np.exp(-multi / 2)
    k /= 2 * np.math.pi *(abs( np.linalg.det(sgm)) ** (1 / 2) ) # linalg.det求行列式的值
    return k


def TraBys(dataMat,i,j,k):
    xMat1, xMat0 = dataMat[0:k], dataMat[k:]
    miu1, miu0, sgm1, sgm0 = getMiuSgm(xMat1, xMat0, i, j)
    return miu1, miu0, sgm1, sgm0


def getMiuSgm(dataMat1, dataMat0, i, j):
    Imiu1 = 0.0
    for r in range(len(dataMat1)):
        Imiu1 += dataMat1[r, i]
    Imiu1 = Imiu1 / len(dataMat1)

    Imiu0 = 0.0
    for r in range(len(dataMat0)):
        Imiu0 += dataMat0[r, i]
    Imiu0 = Imiu0 / len(dataMat0)

    Jmiu1 = 0.0
    for r in range(len(dataMat1)):
        Jmiu1 += dataMat1[r, j]
    Jmiu1 = Jmiu1 / len(dataMat1)

    Jmiu0 = 0.0
    for r in range(len(dataMat0)):
        Jmiu0 += dataMat0[r, j]
    Jmiu0 = Jmiu0 / len(dataMat0)

    miu1 = [Imiu1, Jmiu1]

    miu0 = [Imiu0, Jmiu0]

    sgm1 = np.mat(np.zeros((2, 2)))
    for k in range(len(dataMat1)):
        x = np.mat([dataMat1[k, i] - miu1[0], dataMat1[k, j] - miu1[1]])
        sgm1 += x.T * x
    sgm1 /= len(dataMat1)

    sgm0 = np.mat(np.zeros((2, 2)))
    for k in range(len(dataMat0)):
        x = np.mat([dataMat0[k, i] - miu0[0], dataMat0[k, j] - miu0[1]])
        sgm0 += x.T * x
    sgm0 /= len(dataMat0)
    return miu1, miu0, sgm1, sgm0


def oldShow(i, j, miu1, miu0, sgm1, sgm0):
    if i == 0:
        Ilft = 140
        Irgt = 200
        xlab = 'Height'
    elif i == 1:
        Ilft = 30
        Irgt = 80
        xlab = 'Weight'
    else:
        Ilft = 32
        Irgt = 46
        xlab = 'Shoe Size'
    if j == 0:
        Jlft = 140
        Jrgt = 200
        ylab = 'Height'
    elif j == 1:
        Jlft = 30
        Jrgt = 80
        ylab = 'Weight'
    else:
        Jlft = 32
        Jrgt = 46
        ylab = 'Shoe Size'

    I1 = np.linspace(Ilft, Irgt, 50)
    J1 = np.linspace(Jlft, Jrgt, 50)
    I0 = np.linspace(Ilft, Irgt, 50)
    J0 = np.linspace(Jlft, Jrgt, 50)

    X1, Y1 = np.meshgrid(I1, J1)
    X0, Y0 = np.meshgrid(I0, J0)

    Z1 = []
    for k in range(len(J1)):
        for p in range(len(I1)):
            Z1.append(gauss2([X1[k][p], Y1[k][p]], miu1, sgm1))
    Z1 = np.reshape(Z1, (len(J1), len(I1)))
    Z0 = []
    for k in range(len(J0)):
        for p in range(len(I0)):
            Z0.append(gauss2([X0[k][p], Y0[k][p]], miu0, sgm0))
    Z0 = np.reshape(Z0, (len(J0), len(X0)))

    fig = plt.figure(1)
    ax1 = fig.add_subplot(121, projection='3d')

    ax1.plot_surface(np.array(X1), np.array(Y1), Z1, rstride=1, cstride=1, cmap=plt.cm.coolwarm)

    ax1.plot_surface(np.array(X0), np.array(Y0), Z0, rstride=1, cstride=1, cmap='rainbow')
    plt.title("2D Bayes")

    ax1.set_xlabel(xlab, color='r')
    ax1.set_ylabel(ylab, color='g')
    ax1.set_zlabel('Probability Density', color='b')


def newShow(miu1, miu0, sgm1, sgm0, Ilft, Irgt, Jlft, Jrgt):
    xlab = 'New features of No.0'
    ylab = 'New features of No.1'

    I1 = np.linspace(Ilft, Irgt, 50)
    J1 = np.linspace(Jlft, Jrgt, 50)
    I0 = (Ilft, Irgt, 50)
    J0 = np.linspace(Jlft, Jrgt, 50)

    X1, Y1 = np.meshgrid(I1, J1)
    X0, Y0 = np.meshgrid(I0, J0)

    Z1 = []
    for k in range(len(J1)):
        for p in range(len(I1)):
            Z1.append(gauss2([X1[k][p], Y1[k][p]], miu1, sgm1))
    Z1 = np.reshape(Z1, (len(J1), len(I1)))
    Z0 = []
    for k in range(len(J0)):
        for p in range(len(I0)):
            Z0.append(gauss2([X0[k][p], Y0[k][p]], miu0, sgm0))
    Z0 = np.reshape(Z0, (len(J0), len(I0)))

    fig = plt.figure(1)
    ax2 = fig.add_subplot(122, projection='3d')

    ax2.plot_surface(np.array(X1), np.array(Y1), Z1, rstride=1, cstride=1, cmap=plt.cm.coolwarm)

    ax2.plot_surface(np.array(X0), np.array(Y0), Z0, rstride=1, cstride=1, cmap='rainbow')
    plt.title("PCA-2D Bayes")

    ax2.set_xlabel(xlab, color='r')
    ax2.set_ylabel(ylab, color='g')
    ax2.set_zlabel('Probability Density', color='b')


def newgx(miu1, miu0, sgm1, sgm0, x):
    s1 = (x - miu1) * sgm1.I * (x - miu1).T
    s0 = (x - miu0) * sgm0.I * (x - miu0).T

    s1 = float(s1)
    s0 = float(s0)

    rvs_g1x = s1 + np.log(np.linalg.det(sgm1))
    rvs_g0x = s0 + np.log(np.linalg.det(sgm0))

    return rvs_g0x, rvs_g1x


def TstBys(miu1, miu0, sgm1, sgm0, x, b=0):
    x = np.mat(x)
    miu1 = np.mat(miu1)
    miu0 = np.mat(miu0)

    g1x, g0x = newgx(miu1, miu0, sgm1, sgm0, x)

    if g1x - g0x > b:
        return 1
    else:
        return 0


def getROCxy(dataMat, labelVec, i, j, b, miu1, miu0, sgm1, sgm0):
    FPR = 0.0
    RECALL = 0.0
    FPplusTN = 0.0
    TPplusFN = 0.0

    for k in range(len(labelVec)):

        x = [float(dataMat[k, i]), float(dataMat[k, j])]

        yc = TstBys(miu1, miu0, sgm1, sgm0, x, b)

        if labelVec[k] == 1:
            TPplusFN += 1
        else:
            FPplusTN += 1

        if labelVec[k] == 1 and yc == 1:
            RECALL += 1

        elif labelVec[k] == 0 and yc == 1:
            FPR += 1

    FPR /= FPplusTN

    RECALL /= TPplusFN
    return FPR, RECALL


def oldROC(tstDataMat, i, j, k, old_miu1, old_miu0, old_sgm1, old_sgm0):
    labelVec = [1 if r < k else 0 for r in range(2 * k)]

    X = []
    Y = []

    for b in range(-20, 40, 1):
        FPR, RECALL = getROCxy(tstDataMat, labelVec, i, j, b, \
                               old_miu1, old_miu0, old_sgm1, old_sgm0)

        X.append(FPR)
        Y.append(RECALL)

    plt.plot(X, Y, color='#CC6699', linewidth=2)


def newROC(tstDataMat, k, old_miu1, old_miu0, old_sgm1, old_sgm0):
    labelVec = [1 if r < k else 0 for r in range(2 * k)]

    X = []
    Y = []

    for b in range(-20, 25, 1):
        FPR, RECALL = getROCxy(tstDataMat, labelVec, 0, 1, b, \
                               old_miu1, old_miu0, old_sgm1, old_sgm0)

        X.append(FPR)
        Y.append(RECALL)

    plt.plot(X, Y, color='#9999FF', linewidth=2)


def Gobayes(dataMat, allLowDataMat, k, i=0, j=1):
    lowDataMat = allLowDataMat[0:2 * k]
    oldDataMat = dataMat[0:2 * k]
    newTstDataMat = allLowDataMat[2 * k:]
    tstDataMat = dataMat[2 * k:]

    old_miu1, old_miu0, old_sgm1, old_sgm0 = TraBys(oldDataMat, i, j, k)
    new_miu1, new_miu0, new_sgm1, new_sgm0 = TraBys(lowDataMat, 0, 1, k)

    print("正在绘制类条件概率密度图..")

    oldShow(i, j, old_miu1, old_miu0, old_sgm1, old_sgm0)

    Ilft = min(lowDataMat[:, 0]).tolist()[0][0]
    Irgt = max(lowDataMat[:, 0]).tolist()[0][0]
    Jlft = min(lowDataMat[:, 1]).tolist()[0][0]
    Jrgt = max(lowDataMat[:, 1]).tolist()[0][0]
    newShow(new_miu1, new_miu0, new_sgm1, new_sgm0, Ilft, Irgt, Jlft, Jrgt)
    plt.show()

    print("正在绘制Bayes方法下的ROC曲线..")

    oldROC(tstDataMat, i, j, k, old_miu1, old_miu0, old_sgm1, old_sgm0)
    newROC(lowDataMat, k, new_miu1, new_miu0, new_sgm1, new_sgm0)
    plt.xlabel('FPR')
    plt.ylabel('RECALL')
    plt.title('Bayes\'s ROC curve')
    plt.legend(['2D Bayes', 'PCA-2D Bayes'])
    plt.grid(True)
    plt.show()


def Gobayes2():
    k = 50
    # 获取数据集
    dataMat = getData()
    # 获取降维后的数据集和特征值数组
    allLowDataMat, engVal = MyPCA(dataMat, 2)
    Gobayes(dataMat, allLowDataMat, k)
```
